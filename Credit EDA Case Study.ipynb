### Checking the datatypes and modifying them

```python
curr_app.nunique().sort_values()
```

```python
# changing the data_type of CNT_FAM_MEMBERS as number of family members cannot be in float

curr_app.CNT_FAM_MEMBERS= curr_app.CNT_FAM_MEMBERS.astype('int',errors='ignore')
```

```python
# changing the data_type of DAYS_LAST_PHONE_CHANGE as days cannot be in float

curr_app.DAYS_LAST_PHONE_CHANGE= curr_app.DAYS_LAST_PHONE_CHANGE.astype('int',errors='ignore')
```

```python
# changing the data_type of DAYS_BIRTH as days cannot be in float

curr_app.DAYS_BIRTH= curr_app.DAYS_BIRTH.astype('int',errors='ignore')
```

```python
# changing the data_type of DAYS_EMPLOYED as days cannot be in float

curr_app.DAYS_EMPLOYED= curr_app.DAYS_EMPLOYED.astype('int',errors='ignore')
```

```python
# changing the data_type of DAYS_REGISTRATION as days cannot be in float

curr_app.DAYS_REGISTRATION= curr_app.DAYS_REGISTRATION.astype('int',errors='ignore')
```

```python
# changing the data_type of DAYS_ID_PUBLISH as days cannot be in float

curr_app.DAYS_ID_PUBLISH= curr_app.DAYS_ID_PUBLISH.astype('int',errors='ignore')
```

```python
# Age in days cannot be negative

curr_app.DAYS_BIRTH=curr_app.DAYS_BIRTH.abs()

# Days of employment cannot be negative 

curr_app.DAYS_EMPLOYED=curr_app.DAYS_EMPLOYED.abs()

# DAYS_REGISTRATION cannot be negative

curr_app.DAYS_REGISTRATION=curr_app.DAYS_REGISTRATION.abs()

# DAYS_ID_PUBLISH cannot be negative

curr_app.DAYS_ID_PUBLISH=curr_app.DAYS_ID_PUBLISH.abs()
```

```python
# Calculating applicant's age based on DAYS_BIRTH

curr_app["Age"]=(curr_app.DAYS_BIRTH//365).astype('int')
```

#### CORRELATION for numerical columns

```python
# Correlation of Target=0 data set.

cor_0 = curr_app_0.corr()
cor_0.unstack().reset_index()
corr0 = cor_0.where(np.triu(np.ones(cor_0.shape),k=1).astype(np.bool))
cor_0_df = corr0.unstack().reset_index()
cor_0_df.columns = ['Var1','Var2','Correlation']
cor_0_df.dropna(subset=['Correlation'],inplace=True)
cor_0_df['Correlation_Abs'] = abs(cor_0_df.Correlation)
cor_0_df.sort_values('Correlation_Abs',ascending=False, inplace=True)
cor_0_df
```

```python
# Correlation of Target=1 data set.

cor_1 = curr_app_1.corr()
cor_1.unstack().reset_index()
corr1 = cor_1.where(np.triu(np.ones(cor_1.shape),k=1).astype(np.bool))
cor_1_df = corr1.unstack().reset_index()
cor_1_df.columns = ['Var1','Var2','Correlation']
cor_1_df.dropna(subset=['Correlation'],inplace=True)
cor_1_df['Correlation_Abs'] = abs(cor_1_df.Correlation)
cor_1_df.sort_values('Correlation_Abs',ascending=False, inplace=True)
cor_1_df
```

- People in age group 61-70 having higher credit amounts are likely to fall in defaulter's list.

- People having academic degree and higher credit amount are having higher probability of being defaulter.
- Defaulters are less if credit amount is greater than 2.8 lakh.
- Number of family members being 2 appear to be the best category where there are less chances of default.
- People living on rent or living with parents have higher percentage in Defaulter's list as compared to Non-Defaulter's List.
- People lying in medium slab of credit amount and in low slab of credit amount are at higher risk of being defaulter.
- Very high credit slab people have the less chances of being defaulters and same is applicable to high credit slab people.
